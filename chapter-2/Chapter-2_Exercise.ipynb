## Question 1
What are the two values of the Boolean data type? How do you write them?

## Answer
The two values of the Boolean data type in Python are `True` and `False`. These values are used to represent truth values in logical expressions.

Here's how to write them:

- `True`: Represents a value that is logically true.
- `False`: Represents a value that is logically false.


## Question 2
What are the three Boolean operators?

## Answer
The three Boolean operators in Python are `and`, `or`, and `not`. These operators are used to perform logical operations on Boolean values.

1. `and` Operator:
    - Returns `True` if both operands are `True`.
    - Returns `False` if at least one operand is `False`.


2. `or` Operator:
    - Returns `True` if at least one of the operands is `True`.
    - Returns `False` only if both operands are `False`.

3. `not` Operator:
    - Returns the opposite Boolean value of the operand.
    - If the operand is True, it returns False.
    - If the operand is False, it returns True.

```python
# Here's a more detailed example demonstrating all three operators:

a = True
b = False

# Using 'and' operator
print(a and b)  # Output: False

# Using 'or' operator
print(a or b)   # Output: True

# Using 'not' operator
print(not a)    # Output: False
print(not b)    # Output: True
```
